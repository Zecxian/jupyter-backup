```python
import cloudpickle
with urllib.request.urlopen("http://www.pythonchallenge.com/pc/def/banner.p") as unknown:
    read = cloudpickle.load(unknown)
print(read)
```

```python
for i in read:
    newLine = ""
    for j in i:
        newLine += j[0]*j[1]
    print(newLine)
```
